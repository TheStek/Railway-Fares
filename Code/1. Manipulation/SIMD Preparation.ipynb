```python
import pandas as pd
import numpy as np
```

```python
data_location = r"C:\Users\User\Documents\4. Fourth Year\Project\Railway-Fares\Data\\"
output_location = r"C:\Users\User\Documents\4. Fourth Year\Project\Railway-Fares\Data\Cleansed Data\\"
```

```python
simd = pd.read_csv(data_location + "simd2020_withinds.csv")
simd.head()
```

```python
in_scope_inds = ['Data_Zone', 'Intermediate_Zone', 'Council_area', 'Total_population', 'Working_Age_population','income_count', 'employment_rate', 'employment_count', 'crime_count','crime_rate', 'overcrowded_count', 'nocentralheating_count','overcrowded_rate', 'nocentralheating_rate', 'drive_petrol', 'drive_GP','drive_post', 'drive_primary', 'drive_retail', 'drive_secondary','PT_GP', 'PT_post', 'PT_retail', 'broadband']
simd = simd[in_scope_inds]
```

```python
simd = simd.replace('*', None)
for col in simd.columns:
    if simd[col].dtype == "O":
        if '%' in simd[col][0]:
            print(col)
            simd[col] = simd[col].str.rstrip('%').astype(float) / 100
            
simd["crime_count"] = simd["crime_count"].astype(np.int64)
simd["crime_rate"] = simd["crime_rate"].astype(np.float64)

simd.to_csv(output_location + "simd_dz.csv")
```

```python
aggregations = {
    "Total_population" : "sum",
    "Working_Age_population" : "sum",
    "income_count" : "sum",
    "employment_count" : "sum",
    "crime_count" : "sum",
    "overcrowded_count" : "sum",
    "nocentralheating_count" : "sum",
}
```

```python
simd.groupby("Intermediate_Zone").agg(aggregations)
```
