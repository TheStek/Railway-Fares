```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

from data_helper import *
from GPR_Ensemble import GPR_Ensemble

from sklearn.decomposition import PCA
```

```python
log_name = "gpr_ensemble"
```

```python
from sklearn.gaussian_process.kernels import RBF, Matern, RationalQuadratic, DotProduct
```

```python
kernels = (RBF, Matern, RationalQuadratic, DotProduct)
```

```python
for kernel in kernels:
    for scaling in (True, False):
        check_on_dataset(GPR_Ensemble, "gpr_ensemble", log_name, kernel = kernel(), scaling = scaling)
```

```python
val_scores = []
for alpha in [10**i for i in range(-2, 2)]:
    for l in [10**i for i in range(-2, 2)]:
        model = GPR_Ensemble(kernel = RationalQuadratic(alpha=alpha, length_scale=l))
        model.fit(dataset_in["train"]["dist_remoteness"], dataset_in["train"]["y"])
        val_scores.append((alpha, l, model.score(dataset_in["val"]["dist_remoteness"], dataset_in["val"]["y"])))
```

```python
hyper_df = pd.DataFrame(val_scores, columns = ("alpha", "length scale", "Validation Score"))
hyper_df.sort_values(by = "Validation Score", ascending = False)
```
