```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()

from data_helper import *
```

```python
def test_model(model, subset, model_name, n = 25, sample_size = 0.7):
    def get_sample():
        data = dataset_in["train"][subset].copy()
        data["y"] = dataset_in["train"]["y"]

        data = data.sample(frac = sample_size)

        X_train = data.drop('y', axis = 1)
        y_train = np.ravel(data[['y']])
        
        return X_train, y_train
    
    results = []
    for i in range(n):
        model.fit(*get_sample())
        results.append(model.score(dataset_in["test"][subset], dataset_in["test"]['y']))
    
    df = pd.DataFrame({"Test Score" : results})
    
    df["Model"] = model_name
    df["Subset"] = subset
    df["n"] = n
    
    return df
```

```python
from sklearn.linear_model import LinearRegression

linregs = []
for subset in subsets:
    linregs.append(test_model(scaled_model(LinearRegression), subset, "Linear Regression"))
linreg = pd.concat(linregs)
```

```python
from sklearn.ensemble import RandomForestRegressor

rfrs = []
for subset in subsets:
    rfrs.append(test_model(scaled_model(RandomForestRegressor, max_samples = 0.7), subset, "Random Forest Regression"))
rfr = pd.concat(rfrs)
```

```python
from GPR_Ensemble import GPR_Ensemble
from sklearn.gaussian_process.kernels import RationalQuadratic

gprs = []
for subset in subsets:
    gprs.append(test_model(GPR_Ensemble(kernel = RationalQuadratic(alpha = 0.1), bootstrap = True), subset, "Gaussian Process Regression", n = 10))
gpr = pd.concat(gprs)
```

```python
from sklearn.svm import SVR


svrs = []
for subset in subsets:
    svrs.append(test_model(scaled_model(SVR, kernel = 'rbf', C = 1000, max_iter =1e5, epsilon = 0.5), subset, "Support Vector Regression"))
svr = pd.concat(svrs)
```

```python
all_results = pd.concat([linreg, rfr, gpr, svr])
all_results
```

```python
all_results.to_csv("Test Results.csv", index = False)
```
